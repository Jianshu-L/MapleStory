```python
import numpy as np
```

```python
def magic_attack_range(item_magic, base_int, equip_int, skill_attack, monster_defend, buff):
    if buff:
        int_value = int(base_int*1.1)+equip_int
    else:
        int_value = base_int+equip_int
    magic = item_magic + int_value
    # print(magic)
    max_value = ((magic**2/1000+magic)/30+int_value/200)*1.5*skill_attack-monster_defend*0.5
    min_value = ((magic**2/1000+magic*0.9*0.6)/30+int_value/200)*1.5*skill_attack-monster_defend*0.6

    return (min_value, max_value)
```

```python
import pandas as pd
results = {
    "levels": [],
    "base_int": [],
    "equip_int": [],
    "item_magic": [],
    "buff": [],
    "total_magic": []
    }

for level in range(140,150+1):
    base_int = level*5
    equip_int = 100
    item_magic = 100
    skill_attack = 670

    for add_int in range(200):
        for add_magic in range(200):
            if magic_attack_range(item_magic+add_magic, base_int, equip_int+add_int, skill_attack, 710, True)[0]> 85000:
                results['equip_int'].append(equip_int+add_int)
                results["item_magic"].append(item_magic+add_magic)
                results['levels'].append(level)
                results['base_int'].append(base_int)
                results['buff'].append(True)
                results['total_magic'].append(equip_int+add_int+np.int16(base_int*1.1)+item_magic+add_magic)
                break

for level in range(140,150+1):
    base_int = level*5
    equip_int = 100
    item_magic = 100
    skill_attack = 670

    for add_int in range(200):
        for add_magic in range(200):
            if magic_attack_range(item_magic+add_magic, base_int, equip_int+add_int, skill_attack, 710, False)[0]> 85000:
                results['equip_int'].append(equip_int+add_int)
                results["item_magic"].append(item_magic+add_magic)
                results['levels'].append(level)
                results['base_int'].append(base_int)
                results['buff'].append(False)
                results['total_magic'].append(equip_int+add_int+base_int+item_magic+add_magic)
                break

results['equip_int'] = np.array(results['equip_int'])
results["item_magic"] = np.array(results["item_magic"])
results['levels'] = np.array(results['levels'])
results['base_int'] = np.array(results['base_int'])

data = {
    "level": results['levels'],
    "base_int": results['base_int'],
    "buff": results['buff'],
    "equip_int": results['equip_int'],
    "equip_magic": results["item_magic"],
    "total_magic": results['total_magic']
}
df = pd.DataFrame(data)
df.to_csv("Magic-140-150.csv")
```
