```python
def pop_all_by_value(d: dict, target):
    keys_to_pop = [k for k, v in d.items() if v == target]
    for k in keys_to_pop:
        d.pop(k)
    return keys_to_pop

def pop_by_value_first(d: dict, target):
    for k, v in list(d.items()):  # 用 list 防止遍历时修改字典报错
        if v == target:
            d.pop(k)
            return k, v
    raise KeyError(f"value {target!r} not found")
```

```python
df = pd.read_csv("temp.csv", header=None)
id_today = np.squeeze(df.values).tolist()

today_dict = {}
for i in id_today:
    index_list = find_contain_index(id, i)
    if index_list:
        if len(index_list) != 1:
            raise ValueError("Something Wrong")
        today_dict[id[index_list[0]]] = job_type[index_list[0]]
    else:
        today_dict[i] = []

if len(today_dict) != df.shape[0]:
    raise ValueError("short on people")

nai_list = pop_all_by_value(today_dict, "奶")
quan_list = pop_all_by_value(today_dict, "拳")
yan_list = pop_all_by_value(today_dict, "眼")
huo_list = pop_all_by_value(today_dict, "火")
jinzhan_list = pop_all_by_value(today_dict, "近战")
yuancheng_list = pop_all_by_value(today_dict, "远程")
```

```python
import random
def random_pop_from_list(lst):
    idx = random.randrange(len(lst))  # 随机索引
    item = lst.pop(idx)
    return item
# first team 奶+拳+近战
# second team 奶+眼+远程
# third team 奶+火+贫血怪

num_member = 6
party_member_all = []
party_member_job_all = []

# first team
party_member = []
party_member_job = []
party_member.append(random_pop_from_list(nai_list))
party_member_job.append("奶")
party_member.append(random_pop_from_list(quan_list))
party_member_job.append("拳")
party_member.append(random_pop_from_list(huo_list))
party_member_job.append("火")
for _ in range(3):
    party_member.append(random_pop_from_list(jinzhan_list))
    party_member_job.append("近战")

party_member_all.append(party_member)
party_member_job_all.append(party_member_job)

# second team
party_member = []
party_member_job = []
party_member.append(random_pop_from_list(nai_list))
party_member_job.append("奶")
party_member.append(random_pop_from_list(yan_list))
party_member_job.append("眼")
for _ in range(4):
    if yuancheng_list:
        party_member.append(random_pop_from_list(yuancheng_list))
        party_member_job.append("远程")
    else:
        party_member.append(random_pop_from_list(yan_list))
        party_member_job.append("眼")

party_member_all.append(party_member)
party_member_job_all.append(party_member_job)

all_left = [nai_list, quan_list, yan_list, huo_list, jinzhan_list, yuancheng_list]
all_left_job = ["奶", "拳", "眼", "火", "近战", "远程"]

# third team
party_member = []
party_member_job = []
party_member.append(random_pop_from_list(nai_list))
party_member_job.append("奶")
for i, m_list in enumerate(all_left):
    for m_i in m_list:
        party_member.append(m_i)
        party_member_job.append(all_left_job[i])

party_member_all.append(party_member)
party_member_job_all.append(party_member_job)

# forth team 单挂
party_member_all.append([i for i in today_dict.keys()])

if len(flatten_list(party_member_all)) != df.shape[0]:
    raise ValueError("short on people")
```
